```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Confirmed = pd.read_csv('time_series_covid19_confirmed_global.csv') 
Deaths    = pd.read_csv('time_series_covid19_deaths_global.csv')
Recovered = pd.read_csv('time_series_covid19_recovered_global.csv')
```

```python
Confirmed = Confirmed.T
Deaths    = Deaths.T
Recovered = Recovered.T
```

```python
Confirmed.fillna(0, inplace=True)
Deaths.fillna(0, inplace=True)
Recovered.fillna(0, inplace=True)
```

```python
Confirmed = Confirmed.reset_index()
Deaths    = Deaths.reset_index()
Recovered = Recovered.reset_index()
```

```python
Confirmed.columns = Confirmed.iloc[1]
Deaths.columns    = Deaths.iloc[1]
Recovered.columns = Recovered.iloc[1]
```

```python
Confirmed = Confirmed.iloc[4:,:]
Deaths    = Deaths.iloc[4:,:]
Recovered = Recovered.iloc[4:,:]
```

```python
Confirmed['Date'] = Confirmed['Country/Region']
Deaths['Date'] = Deaths['Country/Region']
Recovered['Date'] = Recovered['Country/Region']
```

```python
del Confirmed['Country/Region']
del Deaths['Country/Region']
del Recovered['Country/Region']
```

```python
Confirmed['Date'] = pd.to_datetime(Confirmed['Date'])
Deaths['Date'] = pd.to_datetime(Deaths['Date'])
Recovered['Date'] = pd.to_datetime(Recovered['Date'])
```

```python
Confirmed = Confirmed.set_index('Date')
Deaths = Deaths.set_index('Date')
Recovered = Recovered.set_index('Date')


Confirmed.index.freq = 'd'
Deaths.index.freq = 'd'
Recovered.index.freq = 'd'
```

```python
Confirmed = Confirmed.groupby(level=0, axis=1).sum()
Deaths = Deaths.groupby(level=0, axis=1).sum()
Recovered = Recovered.groupby(level=0, axis=1).sum()
```

#### Adding Daily basis Data

```python
Confirmed['Daily'] = Confirmed.diff(1).sum(axis = 1)
Deaths['Daily'] = Deaths.diff(1).sum(axis = 1)
Recovered['Daily'] = Recovered.diff(1).sum(axis = 1)
```

```python
# Daily Data
Confirmed_Daily = Confirmed['Daily']
Deaths_Daily    = Deaths['Daily']
Recovered_Daily = Recovered['Daily']


# Global Data
Confirmed_Global = Confirmed.iloc[:,:-1].sum(axis = 1)
Deaths_Global = Deaths.iloc[:,:-1].sum(axis = 1)
Recovered_Global = Recovered.iloc[:,:-1].sum(axis = 1)
```

# <u> TASK [2.a]. Confirmed Cases (Global Total, Global Daily)<u>

```python
print('Globally total Confirmed cases till 2020-07-13 = ',str(int(Confirmed_Global.sum())))
```

```python
Confirmed_Daily.plot(figsize = (12,6));
plt.ylabel('Cases',size = 18)
plt.xlabel('Time Span', size = 18)
plt.title('Global Daily Confirmed Cases', size= 24)
plt.show()
#plt.savefig("out.png")
```

```python
Confirmed_Global.plot(figsize = (12,6));
plt.ylabel('Cases',size = 18)
plt.xlabel('Time Span', size = 18)
plt.title('Global Total Confirmed Cases', size= 24)
plt.show()
```

# <u>TASK [2.b]. Deaths Cases (Global Total, Global Daily)<u>

```python
print('Globally total Deaths till 2020-07-13 = ',str(int(Deaths_Global.sum())))
```

```python
Deaths_Daily.plot(figsize = (12,6));
plt.ylabel('Cases',size = 18)
plt.xlabel('Time Span', size = 18)
plt.title('Global Daily Deaths Cases', size= 24)
plt.show()
```

```python
Deaths_Global.plot(figsize = (12,6));
plt.ylabel('Cases',size = 18)
plt.xlabel('Time Span', size = 18)
plt.title('Global Total Deaths Cases', size= 24)
plt.show()
```

# <u>TASK [2.c]. Recovered Cases (Global Total, Global Daily)<u>

```python
print('Globally total Recovered till 2020-07-13 = ',str(int(Recovered_Global.sum())))
```

```python
Recovered_Daily.plot(figsize = (12,6));
plt.ylabel('Recovered Cases',size = 18)
plt.xlabel('Time Span', size = 18)
plt.title('Global Daily Recoveries', size= 24)
plt.show()
```

```python
Recovered_Global.plot(figsize = (12,6));
plt.ylabel('Recovered Cases',size = 18)
plt.xlabel('Time Span', size = 18)
plt.title('Global Total Recoveries', size= 24)
plt.show()
```

```python
Confirmed_Global.plot.area(stacked = False ,alpha = 0.4, legend = True, label = 'Confirmed', color = 'orange')
Deaths_Global.plot.area(stacked = False ,alpha = 0.4, legend = True, label = 'Deaths', color = 'red')
Recovered_Global.plot.area(stacked = False ,alpha = 0.4,legend = True,label = 'Recovered', figsize = (12,6), color = 'green')
plt.xlabel('Time Span', size = 18)
plt.ylabel('Cases', size = 18)
plt.ylim(0,14000000)
plt.title('Global Bases', size = 24)
plt.show()
```
